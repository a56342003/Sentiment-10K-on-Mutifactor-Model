```python
import pandas as pd
import pickle
```

```python
Rsize_dict = pd.read_pickle('D:/Thesis_data/Rsize_dict.pickle')
Rbtm_dict = pd.read_pickle('D:/Thesis_data/Rbtm_dict.pickle')
```

```python
df = pd.read_excel('D:/Thesis_data/Factors_fama.xlsx')
df.columns = ['Date', 'Rm-Rf', 'SMB', 'HML', 'RF']
df['year'] = df['Date'].map(lambda x: int(str(x)[:4]))
df['month'] = df['Date'].map(lambda x: int(str(x)[4:6]))
df = df[df['year'] != 2005]
df = df[df['year'] < 2019]
df = df.iloc[6:,:].copy().reset_index(drop=True)
df.head()
```

```python
result_list=[]
for year, month_dict in Rsize_dict.items():    
    for month, Rsize in month_dict.items():
        row={}
        row['year_test'] = year
        row['month_test'] = month
        row['SMB_test'] = Rsize
        result_list.append(row)
test_df = pd.DataFrame(result_list)
result_list=[]
for year, month_dict in Rbtm_dict.items():    
    for month, Rbtm in month_dict.items():
        row={}
        row['year_test2'] = year
        row['month_test2'] = month
        row['HML_test'] = Rbtm
        result_list.append(row)
test_df2 = pd.DataFrame(result_list)
test_df = pd.concat([test_df2,test_df],axis = 1).drop(['year_test2','month_test2'], axis=1)
```

```python
final_df = pd.concat([df,test_df], axis=1)
```

```python
def build_dict(df, dict_type):
    result_dict = {}
    for year in range(2005,2020,1):
        result_dict[year] = {}
    for _, row in df.iterrows():
        result_dict[row['year']][row['month']] = row[dict_type]
    return result_dict
```

```python
hml_dict = build_dict(df,'HML')
smb_dict = build_dict(df,'SMB')
```

```python
with open('D:/Thesis_data/hml_dict_fama.pickle', 'wb') as f:
    pickle.dump(hml_dict, f)
with open('D:/Thesis_data/smb_dict_fama.pickle', 'wb') as f:
    pickle.dump(smb_dict, f)
```

```python
rmrf_df = pd.read_excel('D:/Thesis_data/RmRf_fama.xlsx')
rmrf_df.columns = ['date', 'rmrf', 'rf']
rmrf_df['year'] = rmrf_df['date'].map(lambda x: int(str(x)[:4]))
rmrf_df['month'] = rmrf_df['date'].map(lambda x: int(str(x)[4:6]))
```

```python
rmrf_dict = build_dict(rmrf_df, 'rmrf')
rf_dict = build_dict(rmrf_df, 'rf')
```
